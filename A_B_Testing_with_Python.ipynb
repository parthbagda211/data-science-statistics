# **Hypothesis Testing**
Nothing But Assumption Made about something which is later on tested stastically to verify is truth.



1.   **Null Hypothesis(H0)** : we claim that there is no relationship OR difference between groups with respect ot the value of the population paramenter (Treat Everything Same)
2.   **Alternative Hypothesis(H1 or Ha)** : we claim that there is some change, relationship between groups with respect to the value of the population.



# **Steps in Hypothesis Testing**

1.   Formulate (H0) and (Ha)
2.   Decide the Significane level(alpha)
3.   Calculate the test Statistics
4.   Calculate the p-value
5.   Decision to Reject/Retaion the Null Hypothesis



**P-value** : it represents the probabilityof obtaing an observed result OR more extreme result. when the null hypothesis is true.

**Significance Level(alpha)** :  It is a threshol used in hypothesis testion to determine the level of evidence required to reject the Null hypothesis.

**Type-1 error** : we reject the null hypothesis, when in reality it is true.

**Type-2 error** : we accept the null hypothesis, when in reality it is true.

**Decision-Rule :**


1.   **If the p-value <= alpha:** reject the null hypothesis. This means that the observed result is unlikely to have occured by chance alone and there is sufficient evidence to support the alternative hypothesis.
2.   **If the p-value > alpha:** reject the alternative hypothesis.

**t-test :**
 t-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups. It is based on the t-distribution and is applicable when the sample size is small, and the population variance is unknown.

 **There are two main types of t-tests:**


1.   **One-sample t-test:**
This test is used to compare the mean of a single sample to a known or hypothesized population mean (μ0).

 Formula:

 t = (x̄ - μ0) / (s / √n)

  where:

  x̄ = sample mean

  μ0 = hypothesized population mean

  s = sample standard deviation

  n = sample size
2.   **Two-sample t-test:**
This test is used to compare the means of two independent samples.

  Formula for equal variances:

  t = (x̄1 - x̄2) / √[(s1^2/n1) + (s2^2/n2)]

  Formula for unequal variances:

  t = (x̄1 - x̄2) / √[(s1^2/n1) + (s2^2/n2)]

  where:

  x̄1, x̄2 = sample means

  s1, s2 = sample standard deviations
  
  n1, n2 = sample sizes




**ANOVA (Analysis of Variance):** ANOVA is a statistical technique used to determine if there are any statistically significant differences between the means of three or more independent groups. It is an extension of the two-sample t-test for more than two groups.

**Formula:**

F = MSB / MSW

where:

MSB (Mean Square Between) = SSB / (k-1)

MSW (Mean Square Within) = SSW / (N-k)

SSB (Sum of Squares Between) = Σ(n_i * (x̄_i - x̄)^2)

SSW (Sum of Squares Within) = Σ(Σ(x_ij - x̄_i)^2)

k = number of groups

N = total number of observations

n_i = number of observations in group i

x̄_i = mean of group i

x̄ = overall mean

**Chi-square test:** The chi-square test is a statistical hypothesis test used to determine if there is a significant difference between the observed frequencies and expected frequencies of a categorical variable. It is applicable to both goodness-of-fit tests and tests of independence/association in contingency tables.

Formula:

χ^2 = Σ[(O - E)^2 / E]

where:

O = observed frequency

E = expected frequency

```python
from inspect import signature
def ab_test(sample1,sample2,Name=None,significance=None):
  t_statistic, p_value = stats.ttest_ind(sample1,sample2)

  confidence_level = 1 - significance



  n1, n2 = len(sample1), len(sample2)
  df = n1 + n2 - 2
  std_err = np.sqrt(((n1 - 1) * np.var(sample1, ddof=1) + (n2 - 1) * np.var(sample2, ddof=1)) / df * (1 / n1 + 1 / n2))
  margin_of_error = t_statistic * std_err
  mean_diff = np.mean(sample1) - np.mean(sample2)

  confidence_interval = (mean_diff - margin_of_error, mean_diff + margin_of_error)


  if p_value<significance:
    print("Decision : ","Reject The Null Hypothesis")
  else:
    print("Decision : " ,"Reject The Alternative Hpyothesis")


  print(f"Tittle: {Name}")
  print(f"t-statistic: {t_statistic}")
  print(f"p-value: {p_value}")
  print(f"Significance Level: {significance}")
  print(f"Confidence level: {confidence_level * 100}%")
  print(f"Confidence interval: {confidence_interval}")



```

**Null Hypothesis (H0​):** There is no difference in Click Through Rates between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​)**: There is a difference in Click Thorugh Rates between the Light Theme and Dark Theme.

```python
ab_test(dark_theme_ctr,light_theme_ctr,'Click Through Rate Test',0.05)
```

**Null Hypothesis (H0​):** There is no difference in Conversion Rates between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​)**: There is a difference in Conversion Rates between the Light Theme and Dark Theme.

```python
ab_test(dark_theme_cr,light_theme_cr,"Conversion Rate Test",0.05)
```

**Null Hypothesis (H0​):** There is no difference in Bounce Rates between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​):** There is a difference in Bounce Rates between the Light Theme and Dark Theme

```python
ab_test(dark_theme_bounce_rate,light_theme_bounce_rate,"Bounce Rate Test",0.05)
```

**Null Hypothesis (H0​):** There is no difference in Session Duration between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​):** There is a difference in Session Duration between the Light Theme and Dark Theme

```python
ab_test(dark_theme_session_duration,light_theme_session_duration,"Session Duration Test",0.05)
```

# **Categorical Two Sample Test**

```python
from scipy.stats import chi2_contingency
```

```python
def ab_test_cat(sample1,sample2,Name=None,significance=None):
  contingency_table = [[sample1.value_counts()[0], sample1.value_counts()[1]],[sample2.value_counts()[0], sample2.value_counts()[1]]]
  chi2_statistic, p_value, _, _ = chi2_contingency(contingency_table)
  n1 = sum(contingency_table[0])
  n2 = sum(contingency_table[1])
  p1 = contingency_table[0][0] / n1
  p2 = contingency_table[1][0] / n2
  p_pooled = (contingency_table[0][0] + contingency_table[1][0]) / (n1 + n2)
  se_diff = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n1 + 1 / n2))
  z_score = np.abs(stats.norm.ppf(1 - significance / 2))  # Two-tailed confidence level
  margin_of_error = z_score * se_diff

  confidence_level = (1 - significance) * 100
  confidence_interval = (p1 - p2 - margin_of_error, p1 - p2 + margin_of_error)

  if p_value<significance:
    print("Decision : ","Reject The Null Hypothesis")
  else:
    print("Decision : " ,"Reject The Alternative Hpyothesis")


  print(f"Tittle: {Name}")
  print(f"t-statistic: {chi2_statistic}")
  print(f"p-value: {p_value}")
  print(f"Significance Level: {significance}")
  print(f"Confidence level: {confidence_level * 100}%")
  print(f"Confidence interval: {confidence_interval}")

```

**Null Hypothesis (H0​):** There is no difference in Purchases Rates between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​):** There is a difference in Purchases Rates between the Light Theme and Dark Theme

```python
ab_test_cat(dark_theme_purchases,light_theme_purchases,"Purchases Test",0.05)
```

**Null Hypothesis (H0​):** There is no difference in Add to cart Rates between the Light Theme and Dark Theme.

**Alternative Hypothesis (Ha​):** There is a difference in Add to cart Rates between the Light Theme and Dark Theme

```python
ab_test_cat(dark_theme_add2cart,light_theme_add2cart,"Added to cart",0.05)
```
